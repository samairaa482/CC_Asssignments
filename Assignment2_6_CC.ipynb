```python
#Q3

import random

# Generate list
random_numbers = [random.randint(100, 900) for _ in range(100)]

# Check prime function
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

# Count odd, even, and prime numbers
odd = [num for num in random_numbers if num % 2 != 0]
even = [num for num in random_numbers if num % 2 == 0]
prime = [num for num in random_numbers if is_prime(num)]

print("Total odd numbers:", len(odd))
print("Total even numbers:", len(even))
print("Total prime numbers:", len(prime))
```

```python
# Q4

A = {34, 56, 78, 90}
B = {78, 45, 90, 23}

# i. Union
print("Unique scores (Union):", A.union(B))

# ii. Intersection
print("Common scores (Intersection):", A.intersection(B))

# iii. Symmetric Difference
print("Exclusive scores (Symmetric Difference):", A.symmetric_difference(B))

# iv. Subset and Superset check
print("Is A subset of B?", A.issubset(B))
print("Is B superset of A?", B.issuperset(A))

# v. Remove a score X from A
X = int(input("Enter a score to remove from A: "))
if X in A:
    A.remove(X)
    print("Updated A after removal:", A)
else:
    print(f"{X} is not present in set A.")
```

```python
# Q6

import pandas as pd

data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"],
    "Salary": [50000, 60000, 55000, 45000, 70000],
    "Years_of_Experience": [5, 6, 3, 4, 7],
    "Joining_Date": ["2017-01-25", "2018-03-19", "2019-04-22", "2013-09-14", "2011-12-25"],
    "Bonus": [2000, 2500, 1800, 1500, 3000],
    "Age": [30, 28, 25, 35, 26],
    "Performance_Rating": [4.2, 4.8, 3.9, 4.1, 4.7]
}
employees = pd.DataFrame(data)

print(employees.shape)

print(employees.info())

print(employees.describe())

print(employees.head())
print(employees.tail(3))

print("Average salary:", employees["Salary"].mean())
print("Total bonus:", employees["Bonus"].sum())
print("Youngest age:", employees["Age"].min())
print("Highest rating:", employees["Performance_Rating"].max())

print(employees.sort_values(by="Salary", ascending=False))

def categorize_performance(rating):
    if rating >= 4.5:
        return "Excellent"
    elif 4.0 <= rating < 4.5:
        return "Good"
    else:
        return "Average"

employees["Performance_Category"] = employees["Performance_Rating"].apply(categorize_performance)
print(employees)

print(employees.isnull().sum())

employees.rename(columns={"Employee_ID": "ID"}, inplace=True)

print(employees[(employees["Years_of_Experience"] > 5) & (employees["Department"] == "IT")])

employees["Tax"] = employees["Salary"] * 0.1

employees.to_csv("updated_employees.csv", index=False)

```

```python
# Q2

arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]
print("Reversed Array:", reversed_arr)

from collections import Counter

def most_frequent_info(arr):
    counts = Counter(arr)
    most_common = counts.most_common(1)[0][0]
    indices = np.where(arr == most_common)[0]
    return most_common, indices

x = np.array([1,2,3,4,5,1,2,1,1,1])
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])

value_x, indices_x = most_frequent_info(x)
value_y, indices_y = most_frequent_info(y)

print("x: Most frequent =", value_x, "at indices", indices_x)
print("y: Most frequent =", value_y, "at indices", indices_y)
```

```python
#Q2

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create dataset
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Art'],
    'Score': [88, 92, 75, 85, 90]
}

df = pd.DataFrame(data)

# Bar plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Subject', y='Score', data=df, palette='Set2')

# Add annotations
for index, row in df.iterrows():
    plt.text(index, row.Score + 1, row.Score, color='black', ha="center")

plt.title("Scores in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(axis='y')
plt.show()
```

```python
# Q4

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the dataset
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)

# 1. Total profit line plot
plt.figure(figsize=(8, 5))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o')
plt.title("Total Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

# 2. Multiline plot for all products
plt.figure(figsize=(12, 6))
for column in df.columns[1:-1]:  # Exclude month_number and total_profit
    plt.plot(df['month_number'], df[column], marker='o', label=column)
plt.title("Product Sales Over Time")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for each column total
plt.figure(figsize=(10, 6))
totals = df.sum(numeric_only=True)
totals.drop(['month_number'])  # optionally drop month_number if not meaningful
sns.barplot(x=totals.index, y=totals.values, palette='Set3')
plt.title("Total Sales and Profit by Category")
plt.xlabel("Features")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
