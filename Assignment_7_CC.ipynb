```python
                              ## Q1
# PART I: Randomized Sales Data Generation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Initialize Random Seed
roll_number = 123456  # Replace with your roll number
np.random.seed(roll_number)

# 2. Generate Sales Data
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# 3. Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, columns=categories, index=months)
df



# PART II: Data Manipulation and Analysis



# 1. First 5 rows and summary stats
print(df.head())
print(df.describe())

# 2. Total sales per category and per month
total_by_category = df.sum()
total_by_month = df.sum(axis=1)

# 3. Average sales growth between consecutive months per category
avg_growth = df.pct_change().mean()

# 4. Add Total Sales and Growth Rate columns
df['Total Sales'] = df.sum(axis=1)
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
df

# 5. Apply Discount
if roll_number % 2 == 0:
    df['Electronics'] *= 0.9
else:
    df['Clothing'] *= 0.85
df


# PART III: Visualization



# 1. Monthly sales trends (line plot)
df[categories].plot(figsize=(10, 5), marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.grid()
plt.legend()
plt.show()

# 2. Box plots for each category
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[categories])
plt.title('Sales Distribution by Category')
plt.grid()
plt.show()

```

```python
#Q2

import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

# i. Element-wise absolute
abs_array = np.abs(array)
print("Absolute Array:\n", abs_array)

# ii. Percentiles
flat = array.flatten()
percentile_flat = np.percentile(flat, [25, 50, 75])
percentile_col = np.percentile(array, [25, 50, 75], axis=0)
percentile_row = np.percentile(array, [25, 50, 75], axis=1)

print("\nPercentiles of Flattened Array (25th, 50th, 75th):", percentile_flat)
print("Percentiles per Column:\n", percentile_col)
print("Percentiles per Row:\n", percentile_row)

# iii. Mean, Median, Std Dev
mean_flat = np.mean(flat)
median_flat = np.median(flat)
std_flat = np.std(flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nMean of Flattened Array:", mean_flat)
print("Median of Flattened Array:", median_flat)
print("Std Dev of Flattened Array:", std_flat)

print("\nMean per Column:", mean_col)
print("Median per Column:", median_col)
print("Std Dev per Column:", std_col)

print("\nMean per Row:", mean_row)
print("Median per Row:", median_row)
print("Std Dev per Row:", std_row)
```

```python
#Q3

import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floored = np.floor(a)
ceiled = np.ceil(a)
truncated = np.trunc(a)
rounded = np.round(a)

print("Original array:      ", a)
print("Floored values:      ", floored)
print("Ceiled values:       ", ceiled)
print("Truncated values:    ", truncated)
print("Rounded values:      ", rounded)

```
